```python
import math
import matplotlib.pyplot as plt
```

```python
xs = [x for x in range(1, 100)]
ys = [math.sin(math.pi * n) for n in xs]
plt.figure(figsize=(7,4))
plt.scatter(xs, ys)
plt.show()
```

```python
xs = [x for x in range(1, 100)]
ys = [math.cos(math.pi * n) for n in xs]

plt.figure(figsize=(10,5))
plt.scatter(xs, ys)
plt.show()
```

```python
xs = [x for x in range(1, 100)]
ys = [math.tan(math.pi * n) for n in xs]

plt.figure(figsize=(10,5))
plt.scatter(xs, ys)
plt.show()
```

```python
xs = [x for x in range(1, 100)]
ys = [(3 + 1/n)for n in xs]

plt.figure(figsize=(15,3))
plt.scatter(xs, ys)
plt.show()
```

```python
xs = [x for x in range(1, 100)]
ys = [(3 - 1/n)for n in xs]

plt.figure(figsize=(15,3))
plt.scatter(xs, ys)
plt.show()
```
